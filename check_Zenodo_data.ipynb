```python
import os
import fitsio
import numpy as np
from astropy.table import Table

in_directory = '/scratch/ierez/IGMCosmo/VoidFinder/data/DR16S82_H/reconstructed/'
os.chdir(in_directory)
```

```python
def read_pixel_file(pixel_name):
    f = open(pixel_name,'r')
    pixel_data = np.fromfile(f,dtype=np.float64)
    pixel_data = pixel_data.reshape((len(pixel_data)//5,5))
    f.close()
    return(pixel_data)

def read_map_file(map_name,shape_map):
    with open(map_name,'r') as f:
        map_data = np.fromfile(f,dtype=np.float64).reshape(shape_map)
    return(map_data)

def read_catalog(catalog_name,mpc=True):
    file = fitsio.FITS(catalog_name)
    cat = file[1]
    dict_catalog = {}
    if(mpc):
        dict_catalog["X"],dict_catalog["Y"],dict_catalog["Z"],dict_catalog["R"] = cat["X"][:],cat["\
Y"][:],cat["Z"][:],cat["RADIUS"][:]
    else:
        dict_catalog["RA"],dict_catalog["DEC"],dict_catalog["REDSHIFT"],dict_catalog["R"] = cat["RA\
"][:],cat["DEC"][:],cat["REDSHIFT"][:],cat["RADIUS"][:]
    file.close()
    return(dict_catalog)
```

```python
pixel_name = "pixels_lya_tomography_stripe82.bin"
map_name = "map_lya_tomography_stripe82.bin"
shape_map = (2928,90,417)
void_catalog_mpc_name = "catalog_voids_lya_stripe82_mpc.fits"
void_catalog_radecz_name = "catalog_voids_lya_stripe82_radecz.fits"
protocluster_catalog_mpc_name = "catalog_protoclusters_lya_stripe82_mpc.fits"
protocluster_catalog_radecz_name = "catalog_protoclusters_lya_stripe82_radecz.fits"
```

```python
### Loading of pixels used to create the tomography map                                             
pixels = read_pixel_file(pixel_name)

### Loading of the tomography masked map                                                            
map_3d = read_map_file(map_name,shape_map)

### Loading of the void catalog in the (X,Y,Z) Mpc.h-1 coordinates of the map                       
void_catalog_mpc = read_catalog(void_catalog_mpc_name,mpc=True)

### Loading of the void catalog in the (RA,DEC,Redshift) coordinates (RA and DEC in deg)            
void_catalog_radecz = read_catalog(void_catalog_radecz_name,mpc=False)

### Loading of the protocluster catalog in the (X,Y,Z) Mpc.h-1 coordinates of the map               
protocluster_catalog_mpc = read_catalog(protocluster_catalog_mpc_name,mpc=True)

### Loading of the protocluster catalog in the (RA,DEC,Redshift) coordinates (RA and DEC in deg)    
protocluster_catalog_radecz = read_catalog(protocluster_catalog_radecz_name,mpc=False)
```
